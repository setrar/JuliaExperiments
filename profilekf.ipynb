````julia
function matrix_multiplication_kernel(A, B, C)
    idx = thread_position_in_grid_1d()
    N = 10
    for _ in 1:N
        C[idx] += sin(A[idx]) * exp(B[idx])
    end
    return
end;
````

````julia
Metal.@profile @metal threads=n_threads groups=n_groups matrix_multiplication_kernel(A, B, C)
````
